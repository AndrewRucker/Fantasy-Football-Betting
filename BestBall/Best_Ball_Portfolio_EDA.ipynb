#### All player exposure across Sites

```python
df_Pup1 = df_Pup1.reset_index()
df_Pup3 = df_Pup3.reset_index() 
df_PupSF = df_PupSF.reset_index()
df = df.reset_index()
df1 = df1.reset_index()
df3 = df3.reset_index()
df25 = df25.reset_index()
df4 = df4.reset_index()
all_exp = df.append(df1).append(df3).append(df25).append(df4).append(df_Pup1).append(df_Pup3).append(df_PupSF)
number_of_teams = df['Team'].max() + df1['Team'].max() + df3['Team'].max() + df25['Team'].max() + df4['Team'].max() + df_PupSF['Team'].max() + df_Pup3['Team'].max() + df_Pup1['Team'].max()
dk_df = df.append(df3).append(df25).append(df4)
dk_df = dk_df.reset_index()
dk_df['Team'] = dk_df['Capital'].astype(str) + '-' + dk_df['Team'].astype(str)
dk_num_teams = df['Team'].max() + df3['Team'].max() + df25['Team'].max() + df4['Team'].max()
ud_df = df1.append(df_Pup1).append(df_Pup3).append(df_PupSF)
ud_num_teams = df1['Team'].max() + df_PupSF['Team'].max() + df_Pup3['Team'].max() + df_Pup1['Team'].max()
ud_df['Team'] = ud_df['Contest'].astype(str) + '-' + ud_df['Team'].astype(str)

yee = all_exp.drop(columns=['Round', 'Pick', 'ADP', 'Value', 'Draft Start', 'Cumulative Val', 'Strategy', 'Elite Onesie', 'TEAM', 'Team'])
cap_df = yee.pivot_table(
    index=["Name"], columns="Capital", aggfunc="size", fill_value=0
).reset_index()
cap_df.columns = cap_df.columns.astype(str)
cap_df['Cap'] = (cap_df['3'] * 3 )+ (cap_df['4'] * 4)+ (cap_df['5'] * 5) + (cap_df['25'] * 25)
cap_df = cap_df.rename(columns={'Name' : 'Player'})

avg_pick = all_exp.drop(columns=['Team', 'POS', 'TEAM', 'Round', 'ADP', 'Value', 'Draft Start', 'Cumulative Val', 'Strategy', 'Elite Onesie', 'Capital'])
avg_pick = avg_pick.groupby(['Name'], as_index=False)['Pick'].mean()
avg_pick = avg_pick.rename(columns={'Name' : 'Player'})

avg_val = all_exp.drop(columns=['Team', 'POS', 'TEAM', 'Round', 'ADP', 'Pick', 'Draft Start', 'Cumulative Val', 'Strategy', 'Elite Onesie', 'Capital'])
avg_val = avg_val.groupby(['Name'], as_index=False)['Value'].mean()
avg_val = avg_val.rename(columns={'Name' : 'Player'})


exp_df = all_exp['Name'].value_counts()
exp_df = exp_df.reset_index()

exp_df['Count'] = exp_df['Name']
exp_df['%'] = round((exp_df['Count'] / number_of_teams) * 100, 1)
exp_df = exp_df.drop(columns=['Name'])
exp_df.rename({'index' : 'Player'}, axis='columns', inplace=True)
pos_df = all_exp.drop_duplicates(subset='Name')
pos_df = pos_df.drop(columns=['Team', 'TEAM', 'Round', 'Pick', 'ADP', 'Value', 'Draft Start', 'Cumulative Val', 'Strategy', 'Elite Onesie'])
pos_df.rename({'Name' : 'Player'}, axis='columns', inplace=True)
exp_df = exp_df.merge(pos_df, on='Player', how='left')
exp_df = exp_df.drop(columns=['Capital'])
exp_df = exp_df.merge(cap_df, on='Player', how='left')
exp_df = exp_df.merge(avg_pick, on='Player', how='left')
exp_df = exp_df.merge(avg_val, on='Player', how='left')

QB_exp = exp_df.loc[exp_df['POS'] == 'QB']
RB_exp = exp_df.loc[exp_df['POS'] == 'RB']
WR_exp = exp_df.loc[exp_df['POS'] == 'WR']
TE_exp = exp_df.loc[exp_df['POS'] == 'TE']
```
