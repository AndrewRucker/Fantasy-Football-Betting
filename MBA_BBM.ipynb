```python

def filter_teams_by_player(final_df: pd.DataFrame, player_name: str) -> pd.DataFrame:
    # Find rows that contain the player_name in the 'player' column
    player_rows = final_df[final_df['player'].str.contains(player_name)]

    # Get unique team_ids that contain the player_name
    team_ids_with_player = player_rows['team_id'].unique()

    # Filter the original DataFrame by team_ids that contain the player_name
    filtered_df = final_df[final_df['team_id'].isin(team_ids_with_player)]

    return filtered_df
```

#### Rules Association Creationmy

```python
def filter_rules(rules):
    filtered_rules = []
    for i, rule in rules.iterrows():
        antecedent = rule['antecedents']
        consequent = rule['consequents']
        support = rule['support']
        confidence = rule['confidence']
        lift = rule['lift']
        reversed_rule = (consequent, antecedent, support, confidence, lift)
        if reversed_rule not in filtered_rules:
            filtered_rules.append((antecedent, consequent, support, confidence, lift))
    filtered_rules = pd.DataFrame(filtered_rules, columns=['antecedents', 'consequents', 'support', 'confidence', 'lift'])
    return filtered_rules
```

```python
def player_as_antecedent(antecedents):
    return "Jonathan Taylor" in antecedents
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import fpgrowth
from mlxtend.frequent_patterns import association_rules

transactions = filtered_df.groupby('team_id')['player'].apply(list).values.tolist()

te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
basket = pd.DataFrame(te_ary, columns=te.columns_)

frequent_itemsets = fpgrowth(basket, min_support=0.01, use_colnames=True)
rules = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.05)
filtered_rules = filter_rules(rules)

#rules = rules[rules['antecedents'].apply(player_as_antecedent)]
#rules = rules[rules['antecedents'].apply(lambda x: "Jonathan Taylor" in x)]

print(rules)
```
